# Youtube Video To Audio Conveter

```python
#for downloading youtube video
from pytube import YouTube
```

```python
#path to store downloaded videos
path = 'Videos/'
```

```python
video = "https://www.youtube.com/watch?v=UF8uR6Z6KLc"
```

```python
yt = YouTube(video)
yt.title
```

```python
#download video to given path
d_video = yt.streams.first().download(path)
```

```python
#Video To Audio converter
import moviepy.editor as me
```

```python
clip = me.VideoFileClip('Videos/Steve Jobs 2005 Stanford Commencement Address.3gpp')
```

```python
clip.audio.write_audiofile('Audios/audio.wav')
```
